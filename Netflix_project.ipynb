```python
#removing some columns as per your requirement
netflix_df_copy=netflix_df.copy()
netflix_df_copy.drop(['director', 'cast', 'date_added'],axis=1,inplace=True)
netflix_df_copy
```

```python
# filling the null values with own values
netflix_df_copy['country'].replace(np.nan, 'United States', inplace=True)
netflix_df_copy
```

```python
#Exploratory Analysis and Visualization
#we will now visualize the date to get some meaningful insights.
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(13, 13))
ax = sns.countplot(x='rating', data=netflix_df_copy, hue='rating', dodge=False)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.title("Rating")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating', data=netflix_df_copy, hue='type')
plt.title('Relation between Type and Rating of Movies and Shows seperately')
plt.show()
```

```python
#Relation between Movie and TV show
plt.figure(figsize=(10,8))
sns.countplot(x='type', data=netflix_df_copy, hue='type')
plt.title('Type')
plt.show()
```

```python
# finding the country with most number of titles produced
most_titles_produced = netflix_df_copy.groupby('country').count().sort_values('title', ascending=False).head(5)
most_titles_produced.reset_index(inplace=True)
most_titles_produced
```

```python
plt.bar(most_titles_produced.country, most_titles_produced.title)
plt.show()
#Hence, United States has produced the highest numbers of titles.
```

```python
# Number of Movies vs Tv shows
sns.countplot(x='type', data= netflix_df_copy, hue='type')
plt.show()

#Neflix has higher niumber of Movies than Tv shoes.
```

```python
#what are the top 5 most popular rating on Netflix?

netflix_df_copy_rating = netflix_df_copy['rating'].value_counts()
netflix_df_copy_rating = pd.DataFrame(netflix_df_copy_rating).reset_index()
netflix_df_copy_rating.columns = ['rating','Nbr']
sns.barplot(x = 'rating',y = 'Nbr',hue= 'rating', data=netflix_df_copy_rating.head(5));
plt.show()
```

```python
#another way
rate= netflix_df_copy['rating'].value_counts().head(5)
plt.figure(figsize=(8,6))
rate.plot(kind='bar', color='blue')
plt.title('Top 5 most popular rating on Netflix', size=20)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
genre= netflix_df_copy['listed_in'].value_counts().tail(5)
plt.figure(figsize=(10,4))
genre.plot(kind='bar', color='yellow')
plt.title('5 Least popular genres on Netflix', size=20)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
##We can see in the bar chart below about the top 5 popular Netflix genre.
```

```python
genre= netflix_df_copy['listed_in'].value_counts().head(5)
plt.figure(figsize=(10,4))
genre.plot(kind='bar', color='yellow')
plt.title('Top 5 popular genres on Netflix', size=20)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
#We can see in the bar chart below about the top 5 popular Netflix genre.
```

```python
#Which were the top 5 years in number of titles released?
top5_year= netflix_df_copy.groupby('release_year').count().sort_values('title',ascending=False).head(5)
top5_year.reset_index(inplace=True)    
top5_year
```

```python
plt.bar(top5_year.release_year, top5_year.title)
plt.show()
#Clearly, 2018 was the year when highest number of Netflix titles were released, followed by 2017, 2019, 2016 and 2015.
```
